### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Data Preprocessing 

```python
amazon_data = pd.read_csv("amazondataset.csv",usecols=['product_name', 'retail_price','discounted_price', 'brand'], encoding= 'unicode_escape')
flipkart_data = pd.read_csv("flipkartdataset.csv",usecols=['product_name','retail_price', 'discounted_price', 'brand'], encoding= 'unicode_escape')
```

```python
amazon_data.head()
```

```python
flipkart_data.head()
```

```python
data = pd.concat([amazon_data, flipkart_data],axis=1, keys = ['amazon', 'flipkart'])
```
